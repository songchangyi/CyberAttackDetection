```python
import pandas as pd #数据分析
import numpy as np #科学计算
from pandas import Series,DataFrame
BF_logs_raw = pd.read_csv("./data/BF_datalab_httpd_all.log", header = None, delimiter = " ",error_bad_lines=False)
BF_temp = BF_logs_raw[[3,5,6,7,8,9,10]]
# split timestamp
BF_temp['date'] = BF_temp[3].str.split(":").str.get(0)
BF_temp['hour'] = BF_temp[3].str.split(":").str.get(1)
BF_temp['minute'] = BF_temp[3].str.split(":").str.get(2)
BF_temp['seconds'] = BF_temp[3].str.split(":").str.get(3)
# split protocol
BF_temp['protocol'] = BF_temp[5].str.split(" ").str.get(0)
# split comments
BF_temp['username'] = BF_temp[9].str.split(" ").str.get(0)
BF_temp['password'] = BF_temp[9].str.split(" ").str.get(1)
BF_temp['ip_address'] = BF_temp[9].str.split(" ").str.get(2)
BF_temp['comment'] = BF_temp[9].str.split(" ").str.get(4)
BF_temp['label'] = BF_temp[9].str.split(" ").str.get(3)
# drop innecessary columns
BF_temp = BF_temp.drop([3, 5, 8, 9], axis=1)
# rename & replace
BF_temp.rename(columns={6:'status', 7:'respsize', 10:'timed'}, inplace=True)
BF_temp['date'] = BF_temp['date'].str.replace('[', '')
bf_logs=BF_temp[['ip_address','date','hour','minute','seconds','protocol','username','password','status','respsize','timed','label']]
bf_logs.to_csv('./data/bf.csv')
bf_logs
```

```python
Normal_logs_raw = pd.read_csv("./data/normal.log", header = None, delimiter = " ")
Normal_temp = Normal_logs_raw[[3,5,6,7,8,9,10]]
# split timestamp
Normal_temp['date'] = Normal_temp[3].str.split(":").str.get(0)
Normal_temp['hour'] = Normal_temp[3].str.split(":").str.get(1)
Normal_temp['minute'] = Normal_temp[3].str.split(":").str.get(2)
Normal_temp['seconds'] = Normal_temp[3].str.split(":").str.get(3)
# split protocol
Normal_temp['protocol'] = Normal_temp[5].str.split(" ").str.get(0)
# split comments
Normal_temp['username'] = Normal_temp[9].str.split(" ").str.get(0)
Normal_temp['password'] = Normal_temp[9].str.split(" ").str.get(2)
Normal_temp['ip_address'] = Normal_temp[9].str.split(" ").str.get(3)
Normal_temp['comment'] = Normal_temp[9].str.split(" ").str.get(5)
Normal_temp['label'] = Normal_temp[9].str.split(" ").str.get(4)
# drop innecessary columns
Normal_temp = Normal_temp.drop([3, 5, 8, 9], axis=1)
# rename & replace
Normal_temp.rename(columns={6:'status', 7:'respsize', 10:'timed'}, inplace=True)
Normal_temp['date'] = Normal_temp['date'].str.replace('[', '')
normal_log=Normal_temp[['ip_address','date','hour','minute','seconds','protocol','username','password','status','respsize','timed','label']]
normal_log.to_csv('./data/normal.csv')
normal_log
```
