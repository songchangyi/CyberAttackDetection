### Python libs

```python
import pickle
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
```

### Data import

#### BF attack logs

```python
BF_logs_raw = pd.read_csv("./data/BF_datalab_httpd.log", header = None, delimiter = " ")
BF_temp = BF_logs_raw[[3,5,6,7,8,9,10]]
# split timestamp
BF_temp['date'] = BF_temp[3].str.split(":").str.get(0)
BF_temp['hour'] = BF_temp[3].str.split(":").str.get(1)
BF_temp['minute'] = BF_temp[3].str.split(":").str.get(2)
BF_temp['seconds'] = BF_temp[3].str.split(":").str.get(3)
# split protocol
BF_temp['protocol'] = BF_temp[5].str.split(" ").str.get(0)
# split comments
BF_temp['username'] = BF_temp[9].str.split(" ").str.get(0)
BF_temp['password'] = BF_temp[9].str.split(" ").str.get(1)
BF_temp['ip_address'] = BF_temp[9].str.split(" ").str.get(2)
BF_temp['comment'] = BF_temp[9].str.split(" ").str.get(4)
BF_temp['label'] = BF_temp[9].str.split(" ").str.get(3)
# drop innecessary columns
BF_temp = BF_temp.drop([3, 5, 8, 9], axis=1)
# rename & replace
BF_temp.rename(columns={6:'status', 7:'respsize', 10:'timed'}, inplace=True)
BF_temp['date'] = BF_temp['date'].str.replace('[', '')
BF_temp.head()
```

```python
# Save the BF logs
pickle.dump(BF_temp, open('./data/BF_logs.p', 'wb'))
```
